**Analisis con zoom a transacciones con fraude**

```python
#Dataframe de transacciones Fraudulentas
transacciones_fraudulentas=transacciones[transacciones['isFraud']==1]
transacciones_fraudulentas
```

```python
#Grafica de Barras de fraude
plt.figure(figsize=(10,5))
transacciones_fraudulentas_unicas=pd.DataFrame(transacciones_fraudulentas['type'].value_counts()).reset_index()
transacciones_fraudulentas_unicas.columns=['Tipo De Transaccion','Cantidad']
transacciones_fraudulentas_unicas
grafica_de_barras_fraude=sns.barplot(x='Tipo De Transaccion',y='Cantidad',data=transacciones_fraudulentas_unicas)
grafica_de_barras_fraude.bar_label(grafica_de_barras_fraude.containers[0],fmt='%.0f')
plt.xlabel('Tipo De Transacci√≥n')
plt.title('Tipo De Transacci√≥n Fraudulentas')
```

```python
#varaibles_numericas
varaibles_numericas=transacciones_fraudulentas.select_dtypes(include='number').columns
varaibles_numericas
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=transacciones_fraudulentas,x='type',y='amount')
```

```python
#Horas en la que se realizan las transacciones con fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones_fraudulentas, x='step', bins=50, color='blue')
plt.title('Transacciones Fraudulentas por Hora')
plt.xlabel('Horas')
plt.show()

#Horas en la que se realizan las transacciones sin fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones[transacciones['isFraud']==0], x='step', bins=50, color='red')
plt.title('Transacciones No Fraudulentas por Hora')
plt.show()
```

```python
#oldbalanceOrg en la que se realizan las transacciones con fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones_fraudulentas, x='oldbalanceOrg', bins=50, color='blue')
plt.title('Transacciones Fraudulentas por oldbalanceOrg')
plt.show()

#oldbalanceOrg en la que se realizan las transacciones sin fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones[transacciones['isFraud']==0], x='oldbalanceOrg', bins=50, color='red')
plt.title('Transacciones No Fraudulentas por oldbalanceOrg')
plt.show()
```

```python
#newbalanceOrig en la que se realizan las transacciones con fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones_fraudulentas, x='newbalanceOrig', bins=50, color='blue',kde=True)
plt.title('Transacciones Fraudulentas por oldbalanceOrg')
plt.show()

#newbalanceOrig en la que se realizan las transacciones sin fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones[transacciones['isFraud']==0], x='newbalanceOrig', bins=50, color='red',kde=True)
plt.title('Transacciones No Fraudulentas por oldbalanceOrg')
plt.show()
```

```python
#oldbalanceDest en la que se realizan las transacciones con fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones_fraudulentas, x='oldbalanceDest', bins=50, color='blue',kde=True)
plt.title('Transacciones Fraudulentas por oldbalanceOrg')
plt.show()

#oldbalanceDest en la que se realizan las transacciones sin fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones[transacciones['isFraud']==0], x='oldbalanceDest', bins=50, color='red',kde=True)
plt.title('Transacciones No Fraudulentas por oldbalanceOrg')
plt.show()
```

```python
#Monto en la que se realizan las transacciones con fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones_fraudulentas, x='amount', bins=50, color='blue')
plt.title('Transacciones Fraudulentas por Monto')
plt.xlabel('Horas')
plt.show()

#Monto en la que se realizan las transacciones sin fraude
plt.figure(figsize=(20,10))
sns.histplot(data=transacciones[transacciones['isFraud']==0], x='amount', bins=50, color='red')
plt.title('Transacciones No Fraudulentas por Monto')
plt.show()
```
