```python
#Tipo de transacciones unicas

transacciones_unicas=pd.DataFrame(transacciones['type'].value_counts()).reset_index()
transacciones_unicas.columns=['Tipo De Transaccion','Cantidad']
transacciones_unicas
```

```python
#Grafica valores unicos tipos de transaccion
grafica_de_barras=sns.barplot(x='Tipo De Transaccion',y='Cantidad',data=transacciones_unicas)
grafica_de_barras.bar_label(grafica_de_barras.containers[0],fmt='%.0f')
plt.title('Tipo De Transacción')
plt.xlabel('Tipo De Transacción')
plt.ylabel('Cantidad')
```

```python
#Relacion de variables numericas
transacciones.corr(numeric_only=True)
```

```python
sns.heatmap(transacciones.corr(numeric_only=True),annot=True,fmt='.2f',cmap='crest',linewidths=0.5)
plt.title('Correlación entre variables numéricas')
```

```python
#Distribucion de la variable isFraud
Fraudes=sns.countplot(data=transacciones,x='isFraud')
Fraudes.bar_label(Fraudes.containers[0],fmt='%.0f')
```
